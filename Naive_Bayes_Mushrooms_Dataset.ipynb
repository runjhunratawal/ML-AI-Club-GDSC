```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn import preprocessing
```

```python
df = pd.read_csv("mushrooms.csv")
df
```

```python
df.shape
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
df2 = df.apply(label_encoder.fit_transform)
df2

# Here 0 = Edible Class and 1 = Poisonous Class
```

```python
# Splitting Data into training and testing

y = df2["class"].values
x = df2.drop(["class"],axis=1)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state=False)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Naive Bayes

gnb = GaussianNB()
gnb.fit(x_train,y_train)

print("Accuracy: {}%".format(round(gnb.score(x_test,y_test)*100,2)))
```

```python
metrics.classification_report(y_train,y_test)
```
